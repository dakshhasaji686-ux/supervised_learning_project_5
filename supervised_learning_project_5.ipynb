```python
# Numerical columns
num_cols = df.select_dtypes(include=np.number).columns

# Categorical columns
cat_cols = df.select_dtypes(include="object").columns

# Fill missing values
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

### Interpretation

Numerical missing values are replaced with the median, while categorical missing values are replaced with the most frequent value (mode).

```python
X = df.drop(["completion_status", "final_score"], axis=1)

y_cls = df["completion_status"]
y_reg = df["final_score"]
```

### Interpretation

The independent variables are separated from the target variables for both classification and regression tasks.

```python
le = LabelEncoder()

for col in X.select_dtypes(include="object"):
    X[col] = le.fit_transform(X[col])
```

### Interpretation

Machine learning models require numerical input, so all categorical columns are converted into numeric labels.

```python
scaler = StandardScaler()

X = pd.DataFrame(
    scaler.fit_transform(X),
    columns=X.columns
)
```

### Interpretation

Standardization scales all numerical features to a similar range, helping many machine learning algorithms perform better.

```python
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X, y_cls,
    test_size=0.2,
    random_state=42,
    stratify=y_cls
)

print(X_train_c.shape, X_test_c.shape)
```

### Interpretation

The classification dataset is divided into **80% training** and **20% testing** while preserving the class distribution.

```python
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X, y_reg,
    test_size=0.2,
    random_state=42
)

print(X_train_r.shape, X_test_r.shape)
```

### Interpretation

The regression dataset is divided into training and testing sets for model evaluation on unseen data.

# Part C: Bagging (Bootstrap Aggregating)

Bagging combines predictions from multiple base models trained on different bootstrap samples. It helps reduce variance and improves model stability and accuracy.

In this section, we will:
- Train a base Decision Tree model
- Train a Bagging model
- Compare their performances

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import BaggingClassifier, BaggingRegressor

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
import numpy as np
```

### Interpretation

The required models and evaluation metrics are imported for implementing Bagging in both classification and regression tasks.

```python
dt = DecisionTreeClassifier(random_state=42)

dt.fit(X_train_c, y_train_c)

pred_dt = dt.predict(X_test_c)

print("Accuracy :", accuracy_score(y_test_c, pred_dt))
```

### Interpretation

This is the baseline classifier. Its accuracy will be compared with the Bagging Classifier to measure improvement.

```python
bag_cls = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    random_state=42
)

bag_cls.fit(X_train_c, y_train_c)

pred_bag = bag_cls.predict(X_test_c)

print("Accuracy :", accuracy_score(y_test_c, pred_bag))
```

### Interpretation

Bagging combines predictions from multiple Decision Trees, reducing overfitting and improving generalization.

```python
print(classification_report(y_test_c, pred_bag))
```

### Interpretation

The report provides Precision, Recall, F1-Score, and Accuracy to evaluate the classification model comprehensively.

```python
dtr = DecisionTreeRegressor(random_state=42)

dtr.fit(X_train_r, y_train_r)

pred_dt = dtr.predict(X_test_r)

print("R2 Score :", r2_score(y_test_r, pred_dt))
```

### Interpretation

This Decision Tree Regressor serves as the baseline model for comparison with the Bagging Regressor.

```python
bag_reg = BaggingRegressor(
    estimator=DecisionTreeRegressor(),
    n_estimators=100,
    random_state=42
)

bag_reg.fit(X_train_r, y_train_r)

pred_bag = bag_reg.predict(X_test_r)

print("R2 Score :", r2_score(y_test_r, pred_bag))
```

### Interpretation

Bagging Regressor averages predictions from multiple Decision Trees, resulting in improved prediction accuracy and reduced variance.

```python
print("MAE :", mean_absolute_error(y_test_r, pred_bag))

print("RMSE :", np.sqrt(mean_squared_error(y_test_r, pred_bag)))

print("R2 :", r2_score(y_test_r, pred_bag))
```

### Interpretation

- MAE indicates the average prediction error.
- RMSE penalizes larger prediction errors.
- R² Score shows how well the model explains the variance in the target variable.

```python
comparison = {
    "Model": [
        "Decision Tree Classifier",
        "Bagging Classifier",
        "Decision Tree Regressor",
        "Bagging Regressor"
    ],
    "Score": [
        accuracy_score(y_test_c, dt.predict(X_test_c)),
        accuracy_score(y_test_c, pred_bag if False else bag_cls.predict(X_test_c)),
        r2_score(y_test_r, dtr.predict(X_test_r)),
        r2_score(y_test_r, bag_reg.predict(X_test_r))
    ]
}

pd.DataFrame(comparison)
```

### Interpretation

The comparison table summarizes the performance of the base Decision Tree models and their Bagging counterparts.

Generally, Bagging improves performance by reducing variance and making predictions more robust.

# Part D: Boosting Algorithms

Boosting is an ensemble technique where models are trained sequentially.
Each new model tries to correct the errors made by the previous model, leading to better predictive performance.

In this section, we will implement:
- AdaBoost
- Gradient Boosting
- LightGBM
- XGBoost

```python
from sklearn.ensemble import AdaBoostClassifier, AdaBoostRegressor
```

### Interpretation

AdaBoost combines multiple weak learners into a strong learner by giving higher importance to previously misclassified samples.

```python
ada_cls = AdaBoostClassifier(
    n_estimators=100,
    random_state=42
)

ada_cls.fit(X_train_c, y_train_c)

pred_ada = ada_cls.predict(X_test_c)

print("Accuracy :", accuracy_score(y_test_c, pred_ada))
```

### Interpretation

The AdaBoost Classifier improves classification performance by sequentially focusing on difficult-to-classify observations.

```python
print(classification_report(y_test_c, pred_ada))
```

### Interpretation

Precision, Recall, F1-Score, and Accuracy indicate the overall classification performance.

```python
ada_reg = AdaBoostRegressor(
    n_estimators=100,
    random_state=42
)

ada_reg.fit(X_train_r, y_train_r)

pred_ada = ada_reg.predict(X_test_r)

print("R2 Score :", r2_score(y_test_r, pred_ada))
```

### Interpretation

The AdaBoost Regressor improves prediction accuracy by combining multiple weak regression models.

```python
print("MAE :", mean_absolute_error(y_test_r, pred_ada))

print("RMSE :", np.sqrt(mean_squared_error(y_test_r, pred_ada)))

print("R2 :", r2_score(y_test_r, pred_ada))
```

### Interpretation

Lower MAE and RMSE with a higher R² indicate better regression performance.

```python
print("""
AdaBoost Learning Process

Model 1 → Finds Errors
        ↓
Model 2 → Focuses on Previous Errors
        ↓
Model 3 → Focuses on Remaining Errors
        ↓
Final Prediction = Weighted Combination
""")
```

### Interpretation

Unlike Bagging, AdaBoost trains models sequentially, giving higher weight to previously misclassified samples.

```python
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
```

### Interpretation

Gradient Boosting improves predictions by minimizing the loss function using gradient descent principles.

```python
gb_cls = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
    random_state=42
)

gb_cls.fit(X_train_c, y_train_c)

pred_gb = gb_cls.predict(X_test_c)

print("Accuracy :", accuracy_score(y_test_c, pred_gb))
```

### Interpretation

Gradient Boosting sequentially improves the model by correcting prediction errors from previous trees.

```python
gb_reg = GradientBoostingRegressor(
    n_estimators=100,
    learning_rate=0.1,
    random_state=42
)

gb_reg.fit(X_train_r, y_train_r)

pred_gb = gb_reg.predict(X_test_r)

print("R2 Score :", r2_score(y_test_r, pred_gb))
```

### Interpretation

Gradient Boosting Regressor minimizes prediction errors by fitting new trees on residual errors.

```python
print("MAE :", mean_absolute_error(y_test_r, pred_gb))

print("RMSE :", np.sqrt(mean_squared_error(y_test_r, pred_gb)))

print("R2 :", r2_score(y_test_r, pred_gb))
```

### Interpretation

These metrics evaluate the accuracy and error of the regression model.

```python
for lr in [0.01, 0.1, 0.2]:
    model = GradientBoostingClassifier(
        learning_rate=lr,
        random_state=42
    )

    model.fit(X_train_c, y_train_c)

    print(f"LR={lr} :", accuracy_score(y_test_c, model.predict(X_test_c)))
```

### Interpretation

The learning rate controls how much each tree contributes to the final prediction. A smaller learning rate generally improves generalization but requires more trees.

LightGBM is a fast gradient boosting framework designed for high performance and large datasets.

### Interpretation

The LightGBM Regressor efficiently predicts continuous values while reducing training time.

# Part E: Voting & Staking Ensembles

Voting combines predictions from multiple classifiers, while Stacking uses a meta-model to learn from the predictions of base models.

In this section, we implement:
- Voting Classifier
- Hard Voting vs Soft Voting
- Stacking Classifier
- Stacking Regressor

```python
from sklearn.ensemble import VotingClassifier, StackingClassifier, StackingRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression
```

### Interpretation

Voting and Stacking ensemble methods are imported for combining multiple machine learning models.

```python
hard_vote = VotingClassifier(
    estimators=[
        ('dt', DecisionTreeClassifier(random_state=42)),
        ('ada', AdaBoostClassifier(random_state=42)),
        ('gb', GradientBoostingClassifier(random_state=42))
    ],
    voting='hard'
)

hard_vote.fit(X_train_c, y_train_c)

hard_pred = hard_vote.predict(X_test_c)

print("Hard Voting Accuracy:", accuracy_score(y_test_c, hard_pred))
```

### Interpretation

Hard Voting predicts the class that receives the majority of votes from all classifiers.

```python
soft_vote = VotingClassifier(
    estimators=[
        ('dt', DecisionTreeClassifier(random_state=42)),
        ('ada', AdaBoostClassifier(random_state=42)),
        ('gb', GradientBoostingClassifier(random_state=42))
    ],
    voting='soft'
)

soft_vote.fit(X_train_c, y_train_c)

soft_pred = soft_vote.predict(X_test_c)

print("Soft Voting Accuracy:", accuracy_score(y_test_c, soft_pred))
```

### Interpretation

Hard Voting predicts the class that receives the majority of votes from all classifiers.

```python
soft_vote = VotingClassifier(
    estimators=[
        ('dt', DecisionTreeClassifier(random_state=42)),
        ('ada', AdaBoostClassifier(random_state=42)),
        ('gb', GradientBoostingClassifier(random_state=42))
    ],
    voting='soft'
)

soft_vote.fit(X_train_c, y_train_c)

soft_pred = soft_vote.predict(X_test_c)

print("Soft Voting Accuracy:", accuracy_score(y_test_c, soft_pred))
```

### Interpretation

Soft Voting averages class probabilities from all classifiers and predicts the class with the highest probability.

```python
pd.DataFrame({
    "Voting":["Hard","Soft"],
    "Accuracy":[
        accuracy_score(y_test_c, hard_pred),
        accuracy_score(y_test_c, soft_pred)
    ]
})
```

### Interpretation

The comparison shows which voting strategy performs better on the dataset.

```python
stack_cls = StackingClassifier(
    estimators=[
        ('dt', DecisionTreeClassifier(random_state=42)),
        ('ada', AdaBoostClassifier(random_state=42)),
        ('gb', GradientBoostingClassifier(random_state=42))
    ],
    final_estimator=LogisticRegression()
)

stack_cls.fit(X_train_c, y_train_c)

stack_pred = stack_cls.predict(X_test_c)

print("Stacking Accuracy:", accuracy_score(y_test_c, stack_pred))
```

### Interpretation

The Stacking Classifier combines predictions from multiple base models using Logistic Regression as the meta-model.

```python
stack_reg = StackingRegressor(
    estimators=[
        ('dt', DecisionTreeRegressor(random_state=42)),
        ('ada', AdaBoostRegressor(random_state=42)),
        ('gb', GradientBoostingRegressor(random_state=42))
    ],
    final_estimator=LinearRegression()
)

stack_reg.fit(X_train_r, y_train_r)

stack_pred = stack_reg.predict(X_test_r)

print("R2 Score:", r2_score(y_test_r, stack_pred))
```

### Interpretation

The Stacking Regressor combines multiple regression models and uses Linear Regression as the final estimator.

```python
print("MAE :", mean_absolute_error(y_test_r, stack_pred))
print("RMSE :", np.sqrt(mean_squared_error(y_test_r, stack_pred)))
print("R2 :", r2_score(y_test_r, stack_pred))
```

### Interpretation

These metrics evaluate the prediction accuracy of the Stacking Regressor.

# Part F: Model Evaluation & Comparison

This section evaluates all ensemble models using standard classification and regression metrics and identifies the best-performing model.

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

### Interpretation

The required evaluation metrics are imported for assessing model performance.



```python
models = {
    "Bagging": bag_cls,
    "AdaBoost": ada_cls,
    "Gradient Boosting": gb_cls,
    "LightGBM": lgb_cls,
    "XGBoost": xgb_cls,
    "Stacking": stack_cls
}

cls_result = []

for name, model in models.items():
    pred = model.predict(X_test_c)

    cls_result.append([
        name,
        accuracy_score(y_test_c, pred),
        precision_score(y_test_c, pred),
        recall_score(y_test_c, pred),
        f1_score(y_test_c, pred),
        roc_auc_score(y_test_c, pred)
    ])

cls_result = pd.DataFrame(
    cls_result,
    columns=["Model","Accuracy","Precision","Recall","F1","ROC-AUC"]
)

cls_result
```

### Interpretation

The table compares all classification models using Accuracy, Precision, Recall, F1-Score, and ROC-AUC.

```python
reg_models = {
    "Bagging": bag_reg,
    "AdaBoost": ada_reg,
    "Gradient Boosting": gb_reg,
    "LightGBM": lgb_reg,
    "XGBoost": xgb_reg,
    "Stacking": stack_reg
}

reg_result = []

for name, model in reg_models.items():
    pred = model.predict(X_test_r)

    reg_result.append([
        name,
        mean_absolute_error(y_test_r, pred),
        np.sqrt(mean_squared_error(y_test_r, pred)),
        r2_score(y_test_r, pred)
    ])

reg_result = pd.DataFrame(
    reg_result,
    columns=["Model","MAE","RMSE","R2 Score"]
)

reg_result
```

### Interpretation

The table compares all regression models using MAE, RMSE, and R² Score.

```python
cls_result.sort_values("Accuracy", ascending=False)
```

### Interpretation

The model with the highest accuracy is selected as the best classifier.

```python
print("Best Classification Model:")
print(cls_result.loc[cls_result["Accuracy"].idxmax()])

print("\nBest Regression Model:")
print(reg_result.loc[reg_result["R2 Score"].idxmax()])
```

### Interpretation

The best-performing ensemble models are identified for deployment based on evaluation metrics.
